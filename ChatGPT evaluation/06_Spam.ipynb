```python
import pandas as pd
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score, confusion_matrix
```

```python
df = pd.read_excel('ChatGPT Evaluation v2.0.xlsx', sheet_name='6.Spam')
#df = pd.read_csv(filename, delimiter="\t", header=0)
df.head()
```

```python
df["chatgpt_final"].value_counts()
```

```python
df = df[df["chatgpt_final"] != "None"]
df["annotation"].replace({"not spam": 0, "spam": 1}, inplace=True)
df["chatgpt_final"].replace({"[\"not spam\"]": 0, "[\"spam\"]": 1, "['spam']": 1, "['not spam']": 0}, inplace=True) # replacing chatgpt_final values for metrics-friendly format
df.head()
```

```python
y_true = df["annotation"].tolist()
y_pred = df["chatgpt_final"].tolist()
```

```python
# tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()

def evaluate(y_true, y_pred, average="micro"):
    prec = precision_score(y_true, y_pred, average=average)
    rec = recall_score(y_true, y_pred, average=average)
    f1 = f1_score(y_true, y_pred, average=average)
    acc = accuracy_score(y_true, y_pred)
    return prec, rec, f1, acc
```

```python
def evaluate_and_print(y_true, y_pred, average, show_acc=False):
  prec, rec, f1, acc = evaluate(y_true, y_pred, average)
  if show_acc:
    print("\tAccuracy:", acc)
  if not average:
    average = "per class (class 0 = not spam; class 1 = spam)"
  print("average:", average)
  print("\tPrecision:", prec)
  print("\tRecall:", rec)
  print("\tF1:", f1)
```

```python
for i, avg in enumerate(["micro", "macro", "weighted", None]):
  evaluate_and_print(y_true, y_pred, avg, show_acc=(i==0))
```
