```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('ChatGPT Evaluation v2.0.xlsx', sheet_name='8.TextEntail')

df.head()
```

```python
df = df[df['split'] == 'dev']

df.info()
```

```python
df['label'].value_counts()
```

```python
LABEL_MAP = {
    "entailment": 1,
    "not_entailment": 0,
}

gold_labels = [LABEL_MAP[val.lower()] for val in df.label.values]
chatgpt_preds = [LABEL_MAP[val.lower()] for val in df.chatgpt_raw.values]
```

```python
from sklearn.metrics import classification_report

report_lines = classification_report(y_true=gold_labels, y_pred=chatgpt_preds, digits=4).split('\n')

for line in report_lines:
  print(line) 
```

```python
from collections import Counter

c = Counter(gold_labels)
list(c.values())
```

```python
from scipy.stats import entropy
entropy(list(c.values()))
```
