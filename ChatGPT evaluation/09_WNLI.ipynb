```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('ChatGPT Evaluation v2.0.xlsx', sheet_name='9.WNLI')

df.head()
```

```python
df = df[df['split'] == 'dev']
df['chatgpt_final'] = df['chatgpt_final'].astype(str)

df.info()
```

```python
df.query('chatgpt_raw != chatgpt_final')
```

```python
df['label'].value_counts()
```

```python
df['chatgpt_final'] = df['chatgpt_final'].astype(int)
df['label'] = df['label'].astype(int)

gold_labels = df.label.values
chatgpt_preds = df.chatgpt_final.values
```

```python
from sklearn.metrics import classification_report

report_lines = classification_report(y_true=gold_labels, y_pred=chatgpt_preds, digits=4).split('\n')

for line in report_lines:
  print(line) 
```
