```python
import pandas as pd
```

```python
df = pd.read_excel('ChatGPT Evaluation v2.0.xlsx', sheet_name='20.PolEmo')
```

```python
set(df["chatgpt_final"].tolist())
```

```python
final_map = {'None': 'none',
 '["ambiwalentny"]': 'ambiwalwentny',
 '["amibwalentny"]': 'ambiwalwentny',
 '["negatywny"]': 'negatywny',
 '["negetywny"]': 'negatywny',
 '["negiwny"]': 'negatywny',
 '["neutalny"]': 'neutralny',
 '["neutralny"]': 'neutralny',
 '["pozytywny"]': 'pozytywny',
 "['ambiwalentny']": 'ambiwalwentny',
 "['negatywny']": 'negatywny',
 "['neutralny']": 'neutralny',
 "['pozytywny']": 'pozytywny',
 'negatywny': 'negatywny'}
```

```python
df_filtered = df[df.chatgpt_final != 'None']
```

```python
gold = df_filtered["annotation"].tolist()
system = list(map(lambda x: final_map[x], df_filtered["chatgpt_final"].tolist()))
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(gold, system, average='macro') # f1_macro
```

```python
f1_score(gold, system, average='micro') # accuracy
```
