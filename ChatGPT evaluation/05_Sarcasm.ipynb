```python
import pandas as pd
```

```python
df = pd.read_excel('ChatGPT Evaluation v2.0.xlsx', sheet_name='5.Sarcasm')
```

```python
set(df["chatgpt_final"].tolist())
```

```python
df['chatgpt_final'].value_counts()
```

```python
final_map = {'None': 'none',
 '["funny"]': 'funny',
 '["not funny"]': 'not funny',
 'not funny': 'not funny',
 '["not funny"] ': 'not funny'}
```

```python
df_filtered = df[df.chatgpt_final != 'None']
```

```python
gold = df_filtered["annotation"].tolist()
system = list(map(lambda x: final_map[x], df_filtered["chatgpt_final"].tolist()))
```

```python
from sklearn.metrics import f1_score, accuracy_score
```

```python
accuracy_score(gold, system) # accuracy
```

```python
f1_score(gold, system, average='macro') # f1_macro
```

```python
f1_score(gold, system, average='micro') # f1_micro
```

```python
y_true_not_funny = 0
y_true_funny = 0
for y_true, y_pred in zip(gold, system):
  if y_true != y_pred:
    if y_true == 'not funny':
      y_true_not_funny += 1
    else:
      y_true_funny += 1

print('Number of wrong predictions per y_true class:',
      f'y_true == not funny: {y_true_not_funny}',
      f'y_true == funny: {y_true_funny}',
      sep='\n')
```
