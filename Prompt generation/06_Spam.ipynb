# New Section

```python
from datasets import load_dataset
```

```python
sms_dataset = load_dataset("sms_spam")
sms_dataset
```

```python
texts = sms_dataset["train"]["sms"]
new_texts = list(map(lambda x: x.strip(), texts))
labels = sms_dataset["train"]["label"]
```

```python
def make_question(txt):
  prompt = "Answer should be only python list containing string either \"spam\" or \"not spam\". Content of the SMS: {}.".format(txt.strip())
  return prompt
```

```python
sms_prompts = list(map(make_question, new_texts))
sms_prompts[0]
```

```python
label_mapping = {
    0: "not spam",
    1: 'spam',
}
sms_new_labels = list(map(lambda x: label_mapping[x], labels))
sms_new_labels[:10]
```

```python
with open('sms_spam.tsv', 'w') as f:
  for l in zip(sms_prompts, new_texts, sms_new_labels):
    f.write('\t'.join(l) + '\n')
```
