```python
polemo_texts = []
polemo_labels = []
with open('all.text.test.txt') as f:
  for line in f:
    txt, label = line.strip().split("__label__")
    polemo_texts.append(txt)
    polemo_labels.append(label)
```

```python
def make_question(txt):
  prompt = "Oceń sentyment tego tekstu, używając jednej z czterech etykiet: pozytywny, negatywny, neutralny, ambiwalentny. Wynik przedstaw jako listę w języku Python zawierające wybraną etykietę. Tekst do oceny: {}".format(txt)
  return(prompt)
```

```python
polemo_questions = list(map(make_question, polemo_texts))
```

```python
label_mapping = {
    'meta_minus_m': 'negatywny',
    'meta_plus_m': 'pozytywny',
    'meta_zero': 'neutralny',
    'meta_amb': 'ambiwalentny'
}
polemo_new_labels = list(map(lambda x: label_mapping[x], polemo_labels))
polemo_new_labels[:10]
```

```python
with open('polemo.tsv', 'w') as f:
  for l in zip(polemo_questions, polemo_new_labels):
    f.write('\t'.join(l) + '\n')
```
