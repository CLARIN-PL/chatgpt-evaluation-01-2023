```python
import pandas as pd

df_data = pd.read_csv("wsd_prompts.tsv", sep="\t")
df_data.head()
```

```python
df_data = df_data.drop(columns=[
    "taken",
    "api_text_id",
    "chatgpt_raw",
    "chatgpt_final",
    "chatgpt-explanation",
    "gpt35-answer",
    "gpt35-explanation",
])
df_data.head()
```

```python
import numpy as np

df_data['chatgpt_raw'] = np.nan

df_data['prompt'] = df_data['prompt'].apply(lambda x: x + ' Explain your answer.')

df_data.head()
```
